# Cloud Code Masterclass: From Zero to Production

* [Reference Video](https://www.youtube.com/watch?v=AT4b9kLtQCQ)

## Introduction

* `Cloud Code` is a powerful programming Agent designed to assist with `creating`, `refactoring`, and `managing` codebases. 

* This tutorial will take you from basic installation to advanced workflows involving multi-modal inputs, `Figma` integration, and custom agent `skills`.

**Prerequisites:**
- A terminal environment (`PowerShell`, `Bash`, or `Zsh`).
  
- `Node.js` installed (for the example project).
  
- Basic understanding of command-line operations.

## 1. Installation and Modes of Operation

### Basic Setup

To get started, you interact with Cloud Code directly from your terminal.

| Command | Description | Example |
| :--- | :--- | :--- |
| `cloud` | Launches the Cloud Code agent | `cloud` |
| `cloud -login` | Manually triggers authentication | `cloud -login` |
| `cloud -c` | Continues the previous session | `cloud -c` |
| `cloud --DangerouslySkipPermissions` | **Warning**: Disables all permission prompts | `cloud --DangerouslySkipPermissions` |

### Interaction Modes

Cloud Code has three distinct modes affecting how it handles file permissions. Cycle through them using **Shift+Tab**.

## 2. Practical Workflow: Building a Todo App

### Initial Creation
Start by creating a directory and launching the agent.

```bash
mkdir MyTodo
cd MyTodo
cloud
```

Inside the agent, you can issue natural language commands:
> "Create a Todo app using HTML."

### Controlling the Terminal
You can instruct Cloud Code to run terminal commands for you, or use the "Bang" (`!`) syntax to run them yourself without exiting.

| Action | Syntax | Example |
| :--- | :--- | :--- |
| Run Command (Agent) | Natural Language | "Run npm install" |
| Run Command (User) | `! <command>` | `! open index.html` |
| Background Task | `Ctrl + b` | (While task runs) |
| List Tasks | `-tasks` | `-tasks` |

**Background Task Workflow**:
When running a blocking command like a development server (`npm run dev`), you don't want to freeze the agent.

## 3. Advanced Features: MCP & Visuals

Cloud Code supports the **Model Context Protocol (MCP)** to connect with external tools like Figma.

| Command | Purpose |
| :--- | :--- |
| `-mcp` | Open MCP Manager to install/configure servers (e.g., Figma) |
| `-resume` | Resume a specific past conversation |

### Replicating a Figma Design
1.  **Context**: Install the Figma MCP server via `-mcp`.
2.  **Input**: Provide the Figma file link to Cloud Code.
3.  **Process**: Cloud Code uses the MCP tool to inspect layer details (colors, spacing) and generates pixel-perfect code.

## 4. Customization and Resilience

### Context Management
Managing your context window is crucial for long sessions.

| Command | Function |
| :--- | :--- |
| `-compact` / `Ctrl+o` | Compresses history to save tokens |
| `-clear` | Wipes context completely (files remain) |
| `-rewinds` / `Esc Esc` | Undoes recent changes (Time Travel) |

### Persistence with `.cloud.md`
To enable "Memory" across sessions, create a `.cloud.md` file. Cloud Code reads this every time it starts. Use it for:
- Project specific rules ("Always use TypeScript").
- Style guidelines.
- Deployment instructions.

### Hooks
Automate workflows by triggering commands before or after tool use.

**Example: Auto-Format on Save**
To run `prettier` automatically whenever Cloud Code writes a file:
1.  Run `-hooks`.
2.  Select **Post-Write** constraint.
3.  Command: `echo $JSON_PAYLOAD | jq -r .filepath | xargs prettier --write`

## 5. Agents and Skills

### Agent Skills
Skills are reusable prompt templates or "mini-programs" defined in Markdown.
- **Location**: `~/.cloud/skills/`
- **Definition**: A markdown file with YAML frontmatter.

**Example Skill: Daily Report**
```markdown
---
name: Daily Report
description: Write a summary of today's changes
---
Check git diff and list all modified files...
```

### Sub-Agents
For specialized tasks that require isolation (to avoid polluting main context), use Sub-Agents.

| Feature | Agent Skill | Sub-Agent |
| :--- | :--- | :--- |
| **Context** | Shared with main session | Completely Isolated |
| **Best For** | Routine tasks (Summaries) | Heavy tasks (Code Review, Refactoring) |
| **Lifecycle** | Transient | Independent Session |

## 6. Plugins
Plugins act as "Installers" for bundles of Skills, MCP servers, and configurations.
- **Command**: `-plugin`
- **Marketplace**: Browse and install curated setups (e.g., "Frontend Design" plugin which includes color palettes and specialized design prompts).

# Practice Exercises

1.  **Backgrounding**: Start a simple python server (`python -m http.server`). Background it using `Ctrl+b`. Verify it works with `curl`. Then kill it using the `-tasks` menu.
   
2.  **Hook Configuration**: Create a project-level hook that runs `ls -la` every time a file is created. Check the terminal output to verify it triggers.

3.  **Sub-Agent Isolation**: Create a sub-agent for "Documentation". Ask it to write a README. Notice how its "thinking" process doesn't clutter your main history.

4.  **Mode Switching**: Switch to **Plan Mode**. Ask Cloud Code to "Delete all files". Observe that it only *discusses* the plan but cannot execute the deletion until you switch modes.

5.  **Memory**: Create a `.cloud.md` with the instruction: "Always end responses with 'Happy Coding!'". Restart Cloud Code and say "Hi" to verify it reads the memory.

# Reference Answers

### 1. Backgrounding
```bash
cloud
! python -m http.server
# Press Ctrl+b
! curl http://localhost:8000
-tasks
# Press 'k' to kill the python process
```

### 2. Hook Configuration
1.  Run `-hooks`.
2.  Add New -> Matcher: `write_to_file`.
3.  Hook Command: `ls -la`.
4.  Save as "Project Level".
5.  ask: "Create test.txt".
6.  Observe `ls -la` output in terminal.

### 3. Sub-Agent Isolation
1.  `-agent` -> Create New -> "Docs Bot".
2.  Prompt: "Write a README based on file list."
3.  The sub-agent opens a new window/context. Main context only sees the final result "Sub-agent finished".

### 4. Mode Switching
1.  `Shift+Tab` until "Plan Mode".
2.  "Delete everything." -> Agent: "I plan to delete..." (No action).
3.  `ls` -> Files still exist.

### 5. Memory
1.  `touch .cloud.md`.
2.  Edit file: `Instructions: End every response with "Happy Coding!"`
3.  `exit`
4.  `cloud` -> "Hi" -> Response: "... Happy Coding!"
